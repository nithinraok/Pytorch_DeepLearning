```python
%matplotlib inline
import matplotlib.pyplot as plt

import torch
from torchvision import transforms,datasets,models
import numpy
import fc_loader
from torch import nn
import torch.nn.functional as F
import helper
```

```python
transform = transforms.Compose([transforms.Resize((255,255)),
                               transforms.CenterCrop((224,224)),
                               transforms.ToTensor()])
train_transforms = transforms.Compose([transforms.RandomRotation(30),
                                       transforms.RandomResizedCrop(224),
                                       transforms.RandomHorizontalFlip(),
                                       transforms.ToTensor(),
                                       transforms.Normalize([0.485, 0.456, 0.406], 
                                                            [0.229, 0.224, 0.225])])

train_dataset=datasets.ImageFolder("/Users/nithin_rao/MyFiles/Work/Data/fast.ai/dogscats/train/",transform=train_transforms)
test_dataset=datasets.ImageFolder("/Users/nithin_rao/MyFiles/Work/Data/fast.ai/dogscats/valid/",transform=transform)

trainloader=torch.utils.data.DataLoader(train_dataset,batch_size=32,shuffle=True)
testloader=torch.utils.data.DataLoader(test_dataset,batch_size=32,shuffle=True)
```

```python
data_iter = iter(testloader)

images, labels = next(data_iter)
fig, axes = plt.subplots(figsize=(10,4), ncols=4)
for ii in range(4):
    ax = axes[ii]
    helper.imshow(images[ii], ax=ax, normalize=False)
```

```python
torch.cuda.is_available()
```

```python
torch.device("cpu")
```

```python
model = models.densenet121(pretrained=True)
```

```python
print(model)
```

```python
for param in model.parameters():
    param.requires_grad=False
```

```python
from collections import OrderedDict

classifier=nn.Sequential(OrderedDict([
                          ('fc1', nn.Linear(1024, 500)),
                          ('relu', nn.ReLU()),
                          ('fc2', nn.Linear(500, 2)),
                          ('output', nn.LogSoftmax(dim=1))
                          ]))
```

```python
model.classifier=classifier
```

```python
criterion = nn.NLLLoss()
optimizer=torch.optim.Adam(model.classifier.parameters())
```

```python
running_loss=0
for images,labels in trainloader:
    outputs=model.forward(images);
    optimizer.zero_grad()
    loss=criterion(outputs,labels)
    loss.backward()
    optimizer.step()
    running_loss+=loss;
    
    print(loss)
```
