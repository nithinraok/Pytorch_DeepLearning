```python
imgs=!ls ../data/flower_data/train/3/* | head 
```

```python
per=0.9

fig=plt.figure(figsize=(20,4))
for idx,file in enumerate(imgs):
    img=cv2.imread(file);
    ax=fig.add_subplot(2,20/2,idx+1,xticks=[],yticks=[])
    plt.imshow(img)
    aa=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY);
    per_90=sorted(aa.ravel(),reverse=True)[int(np.ceil(per*aa.size))]
    img[aa<per_90]=0
    img=cv2.medianBlur(img,3)
    ax=fig.add_subplot(2,20/2,idx+1+10,xticks=[],yticks=[])
    plt.imshow(img)
```
